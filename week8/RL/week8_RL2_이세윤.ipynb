# **강화학습 설계해보기**

**주제: 주식 거래 인공지능시스템**   
실제로 JP모건이 고객 주문을 최적의 가격으로 빠르게 실행하는 인공지능(AI) 시스템 'LOXM'을 실거래에 도입  
주식을 어떻게 사고 팔지를 심층 강화학습(DRL)이라는 기법을 통해 판단

- **Environment**: 주식 시장
- **State**: 현재 갖고 있는 주식과 그의 가격, 현재의 경제 정보, 주식이 상승세인지 하향세인지 등, 각종 지표 (이 외에도 다양한 방법으로 상태를 정의할 수 있을 것 같다!)
- **Action**: 주식 매수 또는 주식 매도
- **Reward**: 매수 후 매도로 얻는 수익 (수익률)

**한계**  
인공지능은 과거에 지표들이 어떤 상황일 때 매수하면 수익이 큰지를 분석해서 매매전략을 세우게 될 것이다. 하지만 과거를 이용해서 미래를 예측하는 것은 어려울 것이다. 주식 시장은 계속해서 바뀌기 때문에 비슷한 방향으로 접근한다고 해서 항상 수익을 얻을 수는 없기 때문이다. 실제로 주식시장에 오랜 시간 수익을 내는 개인 투자자들도 매년 공부하여 현재 시장에 맞는 새로운 매매법을 찾는다고 한다.  
따라서 자동매매 시스템도 시장이 변하면 매매 전략이 수정되어야 하기 때문에 쉽지 않을 것으로 보인다.
